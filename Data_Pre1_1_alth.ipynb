```python
import matplotlib.pyplot as plt
width_stats = data['pixels_x'].describe()
height_stats = data['pixels_y'].describe()
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(data['pixels_x'], bins=30, color='blue', alpha=0.7)
plt.title('Pixels X (Width) Distribution')
plt.xlabel('Width (pixels)')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(data['pixels_y'], bins=30, color='green', alpha=0.7)
plt.title('Pixels Y (Height) Distribution')
plt.xlabel('Height (pixels)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
noNeed = ['attribution', 'copyright_license', 'acquisition_day', 'lesion_id', 'patient_id', 
          # 'pixels_x', 'pixels_y',
          'diagnosis_1', 'diagnosis_2', 'diagnosis_3', 'diagnosis_4',
          'diagnosis_5','concomitant_biopsy', 'dermoscopic_type','diagnosis_confirm_type',
         'family_hx_mm', 'fitzpatrick_skin_type','mel_class', 'mel_mitotic_index','mel_thick_mm',
         'mel_type', 'mel_ulcer', 'melanocytic', 'nevus_type', 'personal_hx_mm', 'tbp_tile_type',
         'benign_malignant', 'image_type']
df = data.drop(columns = noNeed)
df
```

```python
df = df.dropna()
# df.isnull().sum()
```

```python
for column in df.columns:
    print(df[column].value_counts())
    print("\n")
```

```python
df.loc[:, 'is_melanoma'] = df['diagnosis'].apply(lambda x: 1 if x == 'melanoma' else 0)
df['sex'] = df['sex'].apply(lambda x: 1 if x == 'male' else 0)
df = df.drop(columns = ['diagnosis'])
df
```

```python
site_counts = df['anatom_site_general'].value_counts()
site_mapping = {site: idx for idx, site in enumerate(site_counts.index)}
print(site_mapping)
df['anatom_site_general'] = df['anatom_site_general'].replace(site_mapping)
df
```
