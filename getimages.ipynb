```python
import pandas as pd
import os
import shutil

df = pd.read_csv('./data/sort.csv')

output_folder = 'imagesData'
os.makedirs(output_folder, exist_ok=True)

for isic_id in df['isic_id']:
    image_filename = f"{isic_id}.jpg"  
    image_path = os.path.join('images', image_filename)
    
    if os.path.isfile(image_path):
        shutil.move(image_path, output_folder)
        print(f"移动文件: {image_filename} 到 {output_folder}")
    else:
        print(f"未找到文件: {image_filename}")
```
